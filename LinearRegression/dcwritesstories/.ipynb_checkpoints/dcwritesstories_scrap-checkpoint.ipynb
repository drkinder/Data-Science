```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import re
from collections import Counter
from nltk.corpus import stopwords
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "http://dcwritesstories.com/story-index/"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
a_stories = soup.findAll('a')[54:-3]
stories = {a['title']:{'url':a['href']} for a in a_stories if a.has_attr('title')}
```

```python
cleanr = re.compile('<.*?>')

def cleanhtml(raw_html):
  cleantext = re.sub(cleanr, '', raw_html)
  return cleantext

def extract_body(p_tag_soup):
    body = []
    for line in list(p_tag_soup)[2:]:
        sline = str(line)
        if 'source:' not in sline:
            body.append(cleanhtml(sline))
        else:
            break
    if 'Prompt:' in body[0]:
        return body[2:]
    return body

def get_words_clean(text):
    pass
    
def get_wordfreq(text):
    unwanted_chars = ['”', '\\', '!', '.', ',', '?', ';', ':', '(', ')', '–', '“']
    wordfreq = {}
    for line in text:
        words = line.split(" ")
        for idx, word in enumerate(words):
            
            # Filter \n -- Often \n joins several words together. Break apart and correctly establish each individual word.
            if "\n" in word:
                temp_list = word.split('\n')
                del words[idx]
                words += temp_list
                
            # Filter ' '
            if ' ' in word:
                temp_list = word.split(' ')
                del words[idx]
                words += temp_list
                
            # Filter Punctuation and \n
            for char in unwanted_chars:
                if char in word:
                    word = word.replace(char, '')
            #word = word.replace('\n', ' ')
                     
            word = word.lower()
            if word not in wordfreq:
                wordfreq[word] = 0
            wordfreq[word] += 1
    return wordfreq
```

```python
for s in stories:
    s_raw = requests.get(stories[s]['url'])
    s_soup = BeautifulSoup(s_raw.text, "html.parser")
    s_body = s_soup.find('div', {'class': 'entry-content'})
    body = extract_body(s_soup.findAll('p'))
    stories[s]['body'] = body
    s_published = s_soup.find('time', {'class': 'entry-date published'})
    
    # A few time elements classes are 'entry-date published updated'
    if s_published is not None:
        published = datetime.datetime.strptime(s_published.get_text(), '%B %d, %Y').date()
    else:
        s_published = s_soup.find('time', {'class': 'entry-date published updated'})
        published = datetime.datetime.strptime(s_published.get_text(), '%B %d, %Y').date()
        
    stories[s]['published'] = published
```

```python
pub_dates = [str(stories[p]['published']) for p in stories]
pub_dates.sort()
pub_dates = np.array(pub_dates)
story_count = np.array(list(range(1, len(pub_dates)+1)))
print(story_count)
```

```python
import matplotlib.ticker as ticker

ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(12))
plt.ylabel("Number of Stories")
plt.plot(pub_dates, story_count, 'o', markersize=1)
plt.xticks(rotation=45)
```

```python
def get_date_delta(dataframe, date):
    date_time_obj = pd.to_datetime(date, format="%Y-%m-%d")
    return (date_time_obj - dataframe['Date'].min()) / np.timedelta64(1, 'D')
```

```python
from sklearn.linear_model import LinearRegression

data = [[pub_dates[i], story_count[i]] for i in range(len(story_count))]

df  = pd.DataFrame(data, columns=['Date', 'Count'])
df['Date'] = pd.to_datetime(df['Date'], format="%Y-%m-%d")
df['Date_delta'] = get_date_delta(df, df['Date'])

X = df['Date_delta'][:, np.newaxis]

reg = LinearRegression(fit_intercept=True).fit(X, story_count)
x_test = pub_dates
#x_test = np.array(["2017-04-06", "2018-04-06", "2019-05-21"])
x_test = np.array([get_date_delta(df, i) for i in x_test])[:, np.newaxis]
print(x_test)
y_test = reg.predict(x_test)
difference = y_test-story_count
print(difference)
skew_mean = np.mean(difference)

date = "2017-04-06"
date_test = np.array([get_date_delta(df, date)])
date_test = date_test[:, np.newaxis]
pred = reg.predict(date_test) + skew_mean
print(pred)
print(skew_mean)

date_test = 0
```
